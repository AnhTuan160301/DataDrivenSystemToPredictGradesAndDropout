```python
def createPivotTable(df):
  df.TenMH = df.TenMH.astype('string')
  df['DiemHP'] = pd.to_numeric(df['DiemHP'], errors='coerce')
  options = [ ' Algorithms & Data Structures']
  result_dataset = df[df['TenMH'].isin(options)]
  result_dataset['DiemHP'] = result_dataset['DiemHP'].astype('string')
  result_dataset['DiemHP']
  Student_Id = result_dataset.MaSV
  result_dataset_2 = df[df['MaSV'].isin(Student_Id)]
  result_dataset_2.NienKhoa = result_dataset_2.NienKhoa.astype('string')
  result_dataset_2['NamNHoc'] = result_dataset_2['NienKhoa'].str[:4]
  result_dataset_2['NamNHoc'] = result_dataset_2['NamNHoc'].astype('int')
  result_dataset_2['NHHK'] = pd.to_numeric(result_dataset_2['NHHK'],errors='coerce')
  result_dataset_2['Temp'] = result_dataset_2['NHHK'] - result_dataset_2['NamNHoc']*10
  result_dataset_2['NamHoc'] = result_dataset_2['Temp']//10 + 1
  result_dataset_2['HKHoc'] = result_dataset_2['Temp']%10
  result_dataset_2[['NamHoc', 'HKHoc']]
  result_dataset_2['HKTT'] = result_dataset_2['Temp'].apply(findSemester)
  pivot_table = pd.pivot_table(result_dataset_2, values = 'DiemHP', index = 'MaSV', columns='TenMH')
  pivot_table_HK = pd.pivot_table(result_dataset_2, values = 'HKTT', index = 'MaSV', columns = 'TenMH')
  return pivot_table, pivot_table_HK
```

```python
def createRNNGradeMatrix(pivot_result_HK, pivot_result):
  semesterDict = pivot_result_HK.to_dict(orient = 'index')
  gradeDict = pivot_result.to_dict(orient = 'index')
  dataX = []
  dataY = []
  maxSemester = int(max(pivot_result_HK.max()))
  for id, value in semesterDict.items():
    dataY.append(gradeDict.get(id).get(' Algorithms & Data Structures'))
    temp_List = []
    value.pop(' Algorithms & Data Structures')
    temp_dict = gradeDict.get(id)
    del temp_dict[' Algorithms & Data Structures']
    for k in range(1, maxSemester + 1):
      temporal_dict = temp_dict
      key_list = [key for key, val in value.items() if int(val) == k]
      semesterGradeList = {key : temp_dict[key] for key in key_list}
      temporal_dict = {x: 0.1 for x in temp_dict}
      temporal_dict.update(semesterGradeList)
      grade_list = temporal_dict.values()
      temp_List.append(list(grade_list))
    dataX.append(np.array(temp_List).T)
  return dataX, dataY
```

```python

from keras.utils import to_categorical
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.utils import resample
from sklearn.utils import shuffle
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
from imblearn.combine import SMOTETomek

dataX, dataY = createRNNGradeMatrix(pivot_result_HK, pivot_result)
dataX = np.array(dataX)
dataY = np.array(dataY)
features = int(max(pivot_result_HK.max()))
maxSampleClass = int(max(pivot_result[' Algorithms & Data Structures'].value_counts()))
timeSteps = len(dataX[0])
x_train = []
y_train = []
dataX, dataY = shuffle(dataX, dataY, random_state= 2)
x_train, x_val, y_train, y_val = train_test_split(dataX,dataY, test_size = 0.15)
x_train, x_test, y_train, y_test = train_test_split(x_train, y_train, test_size = 0.15)
scaler = StandardScaler()
x_train = np.reshape(x_train, (-1, features))
x_train = scaler.fit_transform(x_train)
x_train = np.reshape(x_train, (-1, timeSteps, features))
overSample = SMOTETomek()
dataX2d = x_train.reshape(x_train.shape[0], -1)
dataX_new, dataY_new = overSample.fit_resample(dataX2d, y_train)
x_train = np.reshape(dataX_new, (-1, dataX.shape[1], dataX.shape[2]))
y_train = to_categorical(dataY_new)
x_val = np.reshape(x_val, (-1, features))
x_val = scaler.transform(x_val)
x_val = np.reshape(x_val, (-1, timeSteps, features))
y_val = to_categorical(y_val)
x_test = np.reshape(x_test, (-1, features))
x_test = scaler.transform(x_test)
x_test = np.reshape(x_test, (-1, timeSteps, features))
y_test = to_categorical(y_test)
globalVars.x_train, globalVars.x_val, globalVars.y_train, globalVars.y_val = x_train, x_val, y_train, y_val
globalVars.x_train.shape


```

```python

def course_influence(y_predict_real, scaler, model, pivot_result, pivot_result_HK):
  dataX, dataY = createRNNGradeMatrix(pivot_result_HK, pivot_result)
  dataX = np.array(dataX)
  dataY = np.array(dataY)
  features = int(max(pivot_result_HK.max()))
  maxSampleClass = int(max(pivot_result[' Algorithms & Data Structures'].value_counts()))
  timeSteps = len(dataX[0])
  x_train = []
  x_train = np.reshape(dataX, (-1, features))
  x_train = scaler.transform(x_train)
  x_train = np.reshape(x_train, (-1, timeSteps, features))
  y_predict = model.predict(x_train)
  y_pre = np.argmax(y_predict, axis = 1)

  influence = 0
  for i, j in zip(y_pre, y_predict_real):
    temp = i - j
    influence = influence + temp
  return influence
```
